# Project Scope

This project is about looking at the likeliness of having postal transactions or agency transactions under 15 minutes waiting time.

From such insights, management can decide to reallocate resources, re-train staff before converting some of the outlets or closing it down.

```python
#File saved at C:\Users\m.azham\Desktop\EDS\Final Project

import pandas as pd
retail=pd.read_excel('C:/Users/m.azham/Desktop/EDS/Final Project/Retail Set 01.xlsx')
retail.columns=['Month','FY Month','FY','Profit Center','Outlet','State','Region','Total Txn','Agency Txn','Postal Txn', 'Agency Txn(%)','Postal Txn(%)','Total QMS','<15min','15-30min','30min-1hr','>1hr', '<15min(%)','15-30min(%)','30min-1hr(%)','>1hr(%)']
retail.head(5)
```

```python
#Encode
from sklearn import preprocessing

df = pd.DataFrame(retail)
enc = preprocessing.LabelEncoder()

X_2 = df.apply(enc.fit_transform)
X_2.head()
```

```python
X = retail.drop(['<15min(%)'], axis=1).values
y = retail['<15min(%)']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train[0:5]
```

```python
clf = LogisticRegression(C=1).fit(X_train, y_train)
print('Retail')
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```
